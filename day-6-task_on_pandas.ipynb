## 18) Create a column that categorizes employees into age bands (Young, Mid, Senior) and count employees per band per department.


```python
def solve(age):
    if pd.isna(age):
        return np.nan
    elif 18 <= age <= 25:
        return 'young'
    elif 26 <= age <= 40:
        return 'mid'
    elif 41 <= age <= 60:
        return 'senior'
    else:
        return np.nan


df['age_bands']=df['age'].apply(solve)
df
```

```python
df.groupby(df['department'])['age_bands'].value_counts()
```

## 25)Create a final quality score per row based on number of valid fields and flag rows below a quality threshold.

```python
def quality_score(row):
    valid_fields = 0
    total_fields = 0
    
    for col in ['age', 'salary', 'performance_rating', 'experience_years']:
        total_fields += 1
        if pd.notna(row[col]) and row[col] > 0:
            valid_fields += 1
            
    return valid_fields / total_fields if total_fields > 0 else np.nan
df['quality_score'] = df.apply(quality_score, axis=1)
quality_threshold = 0.75
df['quality_flag'] = df['quality_score'] < quality_threshold
df[['employee_id', 'quality_score', 'quality_flag']]
```
